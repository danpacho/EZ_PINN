```python
from src.train.sampler.sampler_2d import Sampler2D
from src.train.sampler.sampler_1d import Sampler1D
import matplotlib.pyplot as plt
```

```python
sampler_1d = Sampler1D(num_samples=10, ranges=(0, 1))

samples_1d = sampler_1d.grid_sample()
samples_1d = samples_1d.numpy()
rand_1d = sampler_1d.random_sample()
rand_1d = rand_1d.numpy()

plt.scatter(samples_1d, [0] * len(samples_1d))
plt.scatter(rand_1d, [0] * len(rand_1d))
```

```python
sampler2d = Sampler2D(
    x_ranges=(-1, 1),
    y_ranges=(-1, 1),
    x_num_samples=10,
    y_num_samples=10,
)
x_grid, y_grid = sampler2d.grid_sample()
x_rand, y_rand = sampler2d.random_sample()

plt.scatter(x_grid, y_grid)
plt.scatter(x_rand, y_rand)
```

```python
from src.train.sampler.rect_sampler import RectBoundarySampler


rect_sampler = RectBoundarySampler(
    domain_x=(-1, 1, 10),
    domain_y=(-1, 1, 10),
    boundary_sample_count=(1000, 10),
)

domain, bc = rect_sampler.sample_rect(["b", "l"])

plt.scatter(domain[0], domain[1], s=5)
plt.scatter(bc[0], bc[1], s=5)

plt.legend(["Domain", "Boundary"])

print(domain[0].shape, domain[1].shape)
print(bc[0].shape, bc[1].shape)
```
